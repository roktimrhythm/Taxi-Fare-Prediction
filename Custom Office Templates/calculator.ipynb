```python
print("\nCalculator")


def addi(a,b):
    return a+b
def subi(a,b):
    return a-b
def multi(a,b):
    return a*b
def divi(a,b):
    return a/b

def cal():
    while True:
        print("\nWELCOME TO THE CALCULATOR")
        print("1. Addition")
        print("2. Substraction")
        print("3. Multiplication")
        print("4. Division")

        choice = input("Enter a valid number:")

        if choice in["1","2","3","4","5"]:

            if choice == "5":
                print("Stop and Exit")
                break
            
            num1 = int(input("Enter the 1st number:"))
            num2 = int(input("Enter the 2st number:"))

            if choice == "1":
                print("Result:",addi(num1,num2))

            elif choice == "2":
                print("Result:",subi(num1,num2))
            
            elif choice == "3":
                print("Result:",multi(num1,num2))

            elif choice == "4":
                print("Result:",divi(num1,num2))
        
        else:
            print("Enter a valid number betwween 1-5")

cal()

```

```python
class BankAccount:

    def __init__(self, name, balance):
        self.name = name
        self.balance = balance
    
    def show_balance(self):
        print(f"{self.name}'s balance: {self.balance}")

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"{amount} deposited successfully")
        else:
            print("Invalid deposit amount")

    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient balance")
        else:
            self.balance -= amount
            print(f"{amount} withdrawn successfully")

    def transfer(self, other_account, amount):
        if amount <= 0:
            print("Invalid amount")
        elif amount > self.balance:
            print("Insufficient balance")
        else:
            self.balance -= amount
            other_account.balance += amount
            print("Transfer successful")



acc1 = BankAccount("Rhythm", 50000)
acc2 = BankAccount("Rahim", 30000)


while True:
    print("\n--- Banking System ---")
    print("1. Show Balance")
    print("2. Deposit")
    print("3. Withdraw")
    print("4. Transfer")
    print("5. Exit")

    choice = input("Enter your choice: ")

    if choice == "5":
        print("Exiting system...")
        break

    if choice in ["1", "2", "3", "4"]:
        
        account_choice = input("Select account (1 for Rhythm, 2 for Rahim): ")

        if account_choice == "1":
            acc = acc1
        elif account_choice == "2":
            acc = acc2
        else:
            print("Invalid account")
            continue

        if choice == "1":
            acc.show_balance()

        elif choice == "2":
            amount = int(input("Enter deposit amount: "))
            acc.deposit(amount)

        elif choice == "3":
            amount = int(input("Enter withdraw amount: "))
            acc.withdraw(amount)

        elif choice == "4":
            amount = int(input("Enter transfer amount: "))
            
            if acc == acc1:
                acc.transfer(acc2, amount)
            else:
                acc.transfer(acc1, amount)

    else:
        print("Invalid choice")
```
